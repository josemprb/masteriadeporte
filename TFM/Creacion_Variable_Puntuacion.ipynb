```python
# RESULTADO
def result(df_final, i):
  return 0.3 if df_final.loc[i,'Result'] == 1 else -0.3 if df_final.loc[i,'Result'] == 3 else 0

# PASES TOTALES
def tot_pass(df_final, i):
  if df_final.loc[i,'TotAtt'] <= 20:
    return 0
  elif (df_final.loc[i,'TotAtt'] > 20) & (df_final.loc[i,'TotAtt'] <= 40):
    if df_final.loc[i,'TotCmpPerc'] <= 50:
      return -0.1
    elif (df_final.loc[i,'TotCmpPerc'] > 50) & (df_final.loc[i,'TotCmpPerc'] <= 60):
      return -0.05
    elif (df_final.loc[i,'TotCmpPerc'] > 60) & (df_final.loc[i,'TotCmpPerc'] <= 70):
      return 0
    elif (df_final.loc[i,'TotCmpPerc'] > 70) & (df_final.loc[i,'TotCmpPerc'] <= 80):
      return 0.025
    elif (df_final.loc[i,'TotCmpPerc'] > 80) & (df_final.loc[i,'TotCmpPerc'] <= 90):
      return 0.05
    else:
      return 0.1
  elif (df_final.loc[i,'TotAtt'] > 40) & (df_final.loc[i,'TotAtt'] <= 60):
    if df_final.loc[i,'TotCmpPerc'] <= 50:
      return -0.2
    elif (df_final.loc[i,'TotCmpPerc'] > 50) & (df_final.loc[i,'TotCmpPerc'] <= 60):
      return -0.1
    elif (df_final.loc[i,'TotCmpPerc'] > 60) & (df_final.loc[i,'TotCmpPerc'] <= 70):
      return 0
    elif (df_final.loc[i,'TotCmpPerc'] > 70) & (df_final.loc[i,'TotCmpPerc'] <= 80):
      return 0.05
    elif (df_final.loc[i,'TotCmpPerc'] > 80) & (df_final.loc[i,'TotCmpPerc'] <= 90):
      return 0.1
    else:
      return 0.2
  else:
    if df_final.loc[i,'TotCmpPerc'] <= 50:
      return -0.3
    elif (df_final.loc[i,'TotCmpPerc'] > 50) & (df_final.loc[i,'TotCmpPerc'] <= 60):
      return -0.15
    elif (df_final.loc[i,'TotCmpPerc'] > 60) & (df_final.loc[i,'TotCmpPerc'] <= 70):
      return 0
    elif (df_final.loc[i,'TotCmpPerc'] > 70) & (df_final.loc[i,'TotCmpPerc'] <= 80):
      return 0.075
    elif (df_final.loc[i,'TotCmpPerc'] > 80) & (df_final.loc[i,'TotCmpPerc'] <= 90):
      return 0.15
    else:
      return 0.3

# PASES EN CORTO
def short_pass(df_final, i):
  if df_final.loc[i,'ShortAtt'] <= 10:
    return 0
  elif (df_final.loc[i,'ShortAtt'] > 10) & (df_final.loc[i,'ShortAtt'] <= 25):
    if df_final.loc[i,'ShortCmpPerc'] <= 50:
      return -0.05
    elif (df_final.loc[i,'ShortCmpPerc'] > 50) & (df_final.loc[i,'ShortCmpPerc'] <= 70):
      return -0.025
    elif (df_final.loc[i,'ShortCmpPerc'] > 70) & (df_final.loc[i,'ShortCmpPerc'] <= 80):
      return 0
    elif (df_final.loc[i,'ShortCmpPerc'] > 80) & (df_final.loc[i,'ShortCmpPerc'] <= 90):
      return 0.025
    else:
      return 0.05
  elif (df_final.loc[i,'ShortAtt'] > 25) & (df_final.loc[i,'ShortAtt'] <= 50):
    if df_final.loc[i,'ShortCmpPerc'] <= 50:
      return -0.1
    elif (df_final.loc[i,'ShortCmpPerc'] > 50) & (df_final.loc[i,'ShortCmpPerc'] <= 70):
      return -0.05
    elif (df_final.loc[i,'ShortCmpPerc'] > 70) & (df_final.loc[i,'ShortCmpPerc'] <= 80):
      return 0
    elif (df_final.loc[i,'ShortCmpPerc'] > 80) & (df_final.loc[i,'ShortCmpPerc'] <= 90):
      return 0.05
    else:
      return 0.1
  else:
    if df_final.loc[i,'ShortCmpPerc'] <= 50:
      return -0.2
    elif (df_final.loc[i,'ShortCmpPerc'] > 50) & (df_final.loc[i,'ShortCmpPerc'] <= 70):
      return -0.1
    elif (df_final.loc[i,'ShortCmpPerc'] > 70) & (df_final.loc[i,'ShortCmpPerc'] <= 80):
      return 0
    elif (df_final.loc[i,'ShortCmpPerc'] > 80) & (df_final.loc[i,'ShortCmpPerc'] <= 90):
      return 0.1
    else:
      return 0.2

# PASES A MEDIA DISTANCIA
def mid_pass(df_final, i):
  if df_final.loc[i,'MedAtt'] <= 10:
    return 0
  elif (df_final.loc[i,'MedAtt'] > 10) & (df_final.loc[i,'MedAtt'] <= 25):
    if df_final.loc[i,'MedCmpPerc'] <= 50:
      return -0.05
    elif (df_final.loc[i,'MedCmpPerc'] > 50) & (df_final.loc[i,'MedCmpPerc'] <= 65):
      return -0.025
    elif (df_final.loc[i,'MedCmpPerc'] > 65) & (df_final.loc[i,'MedCmpPerc'] <= 75):
      return 0
    elif (df_final.loc[i,'MedCmpPerc'] > 75) & (df_final.loc[i,'MedCmpPerc'] <= 90):
      return 0.025
    else:
      return 0.05
  else:
    if df_final.loc[i,'MedCmpPerc'] <= 50:
      return -0.1
    elif (df_final.loc[i,'MedCmpPerc'] > 50) & (df_final.loc[i,'MedCmpPerc'] <= 65):
      return -0.05
    elif (df_final.loc[i,'MedCmpPerc'] > 65) & (df_final.loc[i,'MedCmpPerc'] <= 75):
      return 0
    elif (df_final.loc[i,'MedCmpPerc'] > 75) & (df_final.loc[i,'MedCmpPerc'] <= 90):
      return 0.05
    else:
      return 0.1

# PASES LARGOS
def long_pass(df_final, i):
  if df_final.loc[i,'LongAtt'] <= 10:
    return 0
  else:
    if df_final.loc[i,'LongCmpPerc'] <= 50:
      return -0.1
    elif (df_final.loc[i,'LongCmpPerc'] > 65) & (df_final.loc[i,'LongCmpPerc'] <= 75):
      return 0
    elif (df_final.loc[i,'LongCmpPerc'] > 75) & (df_final.loc[i,'LongCmpPerc'] <= 90):
      return 0.05
    else:
      return 0.1

## CALCULATE POINTS
def calculate_points(df_final, i):
  punt_res = result(df_final, i)
  punt_tot = tot_pass(df_final, i)
  punt_short = short_pass(df_final, i)
  punt_mid = mid_pass(df_final, i)
  punt_long = long_pass(df_final, i)
  df_final.at[i, 'Puntuacion'] += np.sum(
      np.multiply(list(df_final.iloc[i][df_att_values.index]), 
                  list(df_att_values.Values)))
  df_final.at[i,'Puntuacion'] += punt_res + punt_tot + punt_short + punt_mid + punt_long
  return df_final.at[i,'Puntuacion']
```

## OTHERS

```python
df_final = df_final.reset_index(drop=True)
for i in range(len(df_final)):
  df_final.at[i, 'Puntuacion'] = calculate_points(df_final, i)
```
